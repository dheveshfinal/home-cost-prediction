## scikit-learn design

```python
#three type of object
#1.Estimator-it estimate some parameter based on dataset.eg imputer
#2.tranformers- takes input and give output based on the learning from fit()
#3.predictor-LinearRegression
```

```python
#feature scaling method
#1.min-max scalling-(value - min)/max - min(scale down the value in a range),sklearn->MinMaxScaler
#2.standardization-(value - mean)/std - sklearn class StandardScalar
```

```python
from sklearn.pipeline import Pipeline#combine different preprocessing steps and model training into a single
from sklearn.preprocessing import StandardScaler# For example, a feature like "age" might range from 0 to 100, while "income" could range from 30,000 to 200,000. Standardizing ensures that all features contribute equally to the distance calculations used in algorithms like KNN, SVM, or gradient descent optimization.
my_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scalar',StandardScaler())
])
```

```python
housing_num_ur=my_pipeline.fit_transform(housing)
```

```python
housing_num_ur
```

## Selecting desired model

```python
housing=strat_train_set.drop("MEDV",axis=1)
housing_label = strat_train_set["MEDV"].copy()
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(housing_num_ur,housing_label)
```

```python
some_data=housing.iloc[:5]
some_labels=housing_label.iloc[:5]
prepaired_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepaired_data)
```

```python
list(some_labels)#compair
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
housing_prediction=model.predict(housing_num_ur)
lin_mse=mean_squared_error(housing_label,housing_prediction)
lin_mse#which is very less so change the model
rmse=np.sqrt(lin_mse)
lin_mse
```

```python
rmse
```

```python
#in desicion tree its giving 0.0 which is overfitting
```

## cross validation


```python
# mean each time we are changeing the model to find better fit
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,housing_num_ur,housing_label,scoring="neg_mean_squared_error",cv=10)
rmse_scores=np.sqrt(-scores)
rmse_scores#decision tree has less error than this so decision tree is better
```
